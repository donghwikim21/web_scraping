```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import csv
import re
import time
import numpy as np

driver = webdriver.Chrome()
driver.get("https://www.realtor.com/realestateandhomes-search/Queens_NY")
driver.find_elements_by_xpath('//div[@id="srp-list"]')


number_of_items= driver.find_element_by_xpath('.//span[@id="search-result-count"]').text
total = re.findall('\d+', number_of_items)
total = int(total[0]+total[1])
last_page= total//44



csv_file = open('queens.csv', 'w', encoding='utf-8', newline='')

items=['Property','Bed', 'Bath', 'Sqft', 'Price']
writer = csv.writer(csv_file, items)


                                                                     




for i in range(last_page):

    information= driver.find_elements_by_xpath('//ul[@class="srp-list-marginless list-unstyled prop-list"]/li')
    
    info_dict={}
    for i in information:
        

        try:
            Property = i.find_element_by_xpath('.//div[@class="property-type"]').text
            print (Property)
            info_dict['Property']= Property
        except:
            info_dict['Property']= np.nan
            print ('No Property')

        try:
            Bed = i.find_element_by_xpath('.//span[@class="data-value meta-beds"]').text
            print (Bed)
            info_dict['Bed'] = Bed
        except:
            info_dict['Bed'] = np.nan
            print ('No Bed')

        try:
            Bath = i.find_element_by_xpath('.//li[@data-label="property-meta-baths"]/span').text
            print (Bath)
            info_dict['Bath'] = Bath
        except:
            info_dict['Bath'] = np.nan
            print('No Bath')

        try:
            Sqft = i.find_element_by_xpath('.//li[@data-label="property-meta-sqft"]/span').text
            print (Sqft)
            info_dict['Sqft'] = Sqft
        except:
            info_dict['Sqft'] = np.nan
            print ('No Sqft')

        try:
            Price = i.find_element_by_xpath('.//span[@class="data-price"]').text
            print (Price)
            info_dict['Price'] = Price
        except:
            info_dict['Price'] = np.nan
            print ('No Price')

        writer.writerow(info_dict.values())

    y = driver.find_element_by_xpath('//a[@class="next"]')
    ActionChains(driver).move_to_element(y).perform()
    y.click
    time.sleep(2)
csv_file.close()
driver.close()
```
